```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
#from PIL import Image
import matplotlib.image as mpimg
import numpy as np

%matplotlib inline
```

```python
im = np.array(mpimg.imread('frcnn2/train/train_0.jpg'), dtype=np.uint8)

fig, ax = plt.subplots(1)
ax.imshow(im)

with open('frcnn2/output.txt', 'r') as file:
    for _ in range(81):
        line = file.readline()
        s = [int(i) for i in line.split(',')[1:]]
        
        if s[4] != 0:
            rect = patches.Rectangle((s[0], s[1]), s[2]-s[0], s[3]-s[1], 
                                  linewidth = 1, edgecolor = 'r', facecolor = 'none')
            ax.add_patch(rect)
plt.show()
```
